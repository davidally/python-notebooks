#### The United States is already in the Northern and Western Hemispheres so we simply need to descend the list by latitude first. North is 90 degrees whilst the equator is 0. 

```python
url="https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv"
df = pd.read_csv(url)
#Getting the top 5
df.sort_values('lat', ascending=False)[0:5]
```

#### What is the northernmost airport in the United States?

##### Answer: Dillant Hopkins Airport

#### Change the parameter to sort by longitude to find the easternmost. Positive values are east of the Prime Meridian whilst negative values are west.

```python
df.sort_values('lon', ascending=True)[0:5]
```

#### What is the easternmost airport in the United States?

##### Answer: Eareckson Air Station

```python
df.sort_values('tzone')
```

#### Which New York area airport had the windiest weather?

```python
df.loc[df['tzone']=='America/New_York'].sort_values('alt', ascending=False)[0:5]
```

##### Answer: Mercer County Airport had the windiest weather due to having the highest altitude. Wind strength increases with higher altitude.
